### Dicionário do Conjunto de Dados (Por Coluna)

- **ID do Estudante:** Um identificador único atribuído a cada estudante.
- **Idade:** A idade dos estudantes varia de 15 a 18 anos.
- **Gênero:** Gênero dos estudantes, onde 0 representa Masculino e 1 representa Feminino.
- **Etnia:** A etnia dos estudantes, codificada da seguinte forma:
 - 0: Caucasiano;
 - 1: Afro-Americano;
 - 2: Asiático;
 - 3: Outro.
- **Educação dos Pais:** O nível de educação dos pais, codificado da seguinte forma:
 - 0: Nenhuma;
 - 1: Ensino Médio;
 - 2: Algum Ensino Superior;
 - 3: Bacharelado;
 - 4: Superior.
- **Tempo de Estudo Semanal:** Tempo de estudo semanal em horas, variando de 0 a 20.
- **Faltas:** Número de faltas durante o ano letivo, variando de 0 a 30.
- **Aulas de Reforço:** Status de aulas de reforço, onde 0 indica Não e 1 indica Sim.
- **Apoio dos Pais:** O nível de apoio dos pais, codificado da seguinte forma:
 - 0: Nenhum;
 - 1: Baixo;
 - 2: Moderado;
 - 3: Alto;
 - 4: Muito Alto.
- **Atividades Extracurriculares:** Participação em atividades extracurriculares, onde 0 indica Não e 1 indica Sim.
- **Esportes:** Participação em esportes, onde 0 indica Não e 1 indica Sim.
- **Música:** Participação em atividades musicais, onde 0 indica Não e 1 indica Sim.
- **Voluntariado:** Participação em voluntariado, onde 0 indica Não e 1 indica Sim.
- **GPA:** Média de Notas (Grade Point Average) em uma escala de 2,0 a 4,0, influenciada por hábitos de estudo, envolvimento dos pais e atividades extracurriculares.
- **Variável Alvo: Classificação de Notas**
- **Classificação de Notas:** Classificação das notas dos estudantes com base no GPA:
 - 0: 'A' (GPA >= 3,5);
 - 1: 'B' (3,0 <= GPA < 3,5);
 - 2: 'C' (2,5 <= GPA < 3,0);
 - 3: 'D' (2,0 <= GPA < 2,5);
 - 4: 'F' (GPA < 2,0).

Indicando que a idade pode não ser um fator importante nas notas dos alunos.
- **Para testar as observações realizadas acima, procede-se, na sequência com o teste de qui-quadrado:**
 - O teste de qui-quadrado é uma técnica estatística utilizada para testar a independência entre duas variáveis categóricas ou para verificar a adequação de uma distribuição observada em relação a uma distribuição esperada.
  - Existem alguns pressupostos que devem ser atendidos para que os resultados do teste de qui-quadrado sejam válidos:
    - Dados Categóricos: O teste de qui-quadrado é adequado para variáveis categóricas (nominais ou ordinais). Os dados devem ser organizados em uma tabela de contingência, onde as frequências observadas são comparadas com as frequências esperadas.
    - Tamanho da Amostra: Geralmente, recomenda-se que cada célula da tabela de contingência tenha uma frequência esperada de pelo menos 5 para garantir a validade do teste.
    - As observações devem ser independentes umas das outras. Ou seja cada observação deve ser coletada de forma independente.
    
- H0 (Hipótese Nula) = Não há diferença significativa entre as frequências observadas e esperadas, o que, por sua vez, sugere que as variáveis são independentes.
- HA (Hipótese Alternativa) = Há diferença significativa entre as frequências observadas e esperadas,ou seja, as variáveis não são independentes, há associação.
- α=0,05: Vamos utilizar o nível de significância mais comumente usado.

```python
# Realizando o teste qui-quadrado para as variáveis categóricas
chi2_age, p_age = chi2_test('Age')

print(f'Chi-quadrado: {chi2_age}')
print(f'P-valor: {p_age:.5f}')

# Interpretação dos resultados categóricos
# Interpretação dos resultados
if p_age < 0.05:
    print("Há uma associação estatisticamente significativa entre a idade e o desempenho acadêmico.")
else:
    print("Não há associação estatisticamente significativa entre a idade e o desempenho acadêmico.")
```

- **O valor de p obtido é maior que o nível de significância estabelecido a priori de 0.05. Diante desse resultado, aceitamos a hipótese nula de que as variáveis são indepedentes. Portanto, com base no valor de p obtido, podemos concluir que não há uma associação estatisticamente significativa entre as variáveis.** Isso indica que, dentro deste conjunto de dados e ao nível de confiança usual de 95%, não foi encontrada uma associação significativa entre as duas variáveis.
- O p-valor indica que as diferenças observadas entre as frequências esperadas e observadas são muito pequenas. Isso sugere uma forte evidência de que não há uma associação significativa entre a idade e o desempenho do estudante ('GradeClass').

 A estratégia será:
   - Usar o Kruskal-Wallis para avaliar a relação entre 'Absences' e 'GradeClass', considerando ser um teste menos sensível a outlier do que o ANOVA, que usa as médias.
   - Isolar os outliers para avaliá-los melhor.




```python
# Aplicando o teste de Kruskal-Wallis para 'Absences' em relação a 'GradeClass'
kruskal_test = kruskal(
  df_student_performance[df_student_performance['GradeClass'] == 0]['Absences'],
  df_student_performance[df_student_performance['GradeClass'] == 1]['Absences'],
  df_student_performance[df_student_performance['GradeClass'] == 2]['Absences'],
  df_student_performance[df_student_performance['GradeClass'] == 3]['Absences'],
  df_student_performance[df_student_performance['GradeClass'] == 4]['Absences']
)

# Exibindo os resultados
print(f"Estatística do teste de Kruskal-Wallis: {kruskal_test.statistic}")
print(f"P-valor: {kruskal_test.pvalue}")

# Interpretação
if kruskal_test.pvalue < 0.05:
    print("Existe uma diferença significativa no número de faltas entre as categorias de GradeClass.")
else:
    print("Não existe uma diferença significativa no número de faltas entre as categorias de GradeClass.")
```

```python
# Função para calcular os outliers por grupo
def find_outliers_by_group(df, group_col, target_col):
  outliers_list = pd.DataFrame()  # DataFrame vazio para armazenar outliers
  unique_groups = df[group_col].unique()  # Identificar grupos únicos (GradeClass)

  for group in unique_groups:
    group_data = df[df[group_col] == group]  # Filtrar os dados por grupo
    Q1 = group_data[target_col].quantile(0.25)  # Primeiro quartil
    Q3 = group_data[target_col].quantile(0.75)  # Terceiro quartil
    IQR = Q3 - Q1  # Intervalo interquartil

    # Definindo os limites para outliers
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Filtrando os outliers dentro do grupo
    outliers = group_data[(group_data[target_col] < lower_bound) | (group_data[target_col] > upper_bound)]

    # Adicionando os outliers ao DataFrame
    outliers_list = pd.concat([outliers_list, outliers])

  return outliers_list

# Encontrando outliers por grupo 'GradeClass' para a variável 'Absences'
outliers_by_group_Absences = find_outliers_by_group(df_student_performance, 'GradeClass', 'Absences')

# Exibindo os dados dos outliers
outliers_by_group_Absences
```

```python
# Contando o número de outliers
num_outliers_Absences = len(outliers_by_group_Absences)

# Exibindo o número de outliers
print(f"Número de outliers encontrados: {num_outliers_Absences}")
```

# Aplicação de Modelos de Aprendizagem

* Os três modelos de aprendizagem escolhidos para lidar com este dataset, foram:
 * Random Forest Classifier: pois é um modelo robusto, que lida bem com dados desbalanceados e indicado para identificar a importância entre as variáveis. Além disso, pode lidar com variáveis categóricas e contínuas.
 * Gradient Boosting (XGBoost): é um modelo poderoso, que, frequentemente, supera outros modelos, quando se trata de problemas tabulares. Também lida com dados desbalanceados, sendo eles categóricos e/ou contínuos.
 * Logistic Regression: dada sua simplicidade de interpretação, ideal para problemas de classificação binária ou multiclass. Funciona bem com correlação linear entre as variáveis preditoras e a variável-alvo.

```python
# Remover as variáveis irrelevantes conforme indicado
df = df_student_performance.drop(columns=['Age', 'Gender', 'Ethnicity', 'ParentalEducation', 'StudentID'])
```

```python
# Dividir entre X (features) e y (variável-alvo)
X = df.drop(columns=['GradeClass'])
y = df['GradeClass']
```

```python
# Definir pré-processamento: normalização para variáveis contínuas e one-hot encoding para categóricas
numeric_features = ['StudyTimeWeekly', 'Absences', 'GPA']
categorical_features = ['ParentalSupport', 'Extracurricular', 'Sports', 'Music', 'Volunteering']
```

```python
preprocessor = ColumnTransformer(
  transformers=[
      ('num', StandardScaler(), numeric_features),
      ('cat', OneHotEncoder(), categorical_features)])
```

```python
# Modelo 1: Random Forest
pipeline_rf = Pipeline(steps=[('preprocessor', preprocessor),
                              ('classifier', RandomForestClassifier(random_state=42))])
```

```python
# Modelo 2: XGBoost
pipeline_xgb = Pipeline(steps=[('preprocessor', preprocessor),
                               ('classifier', XGBClassifier(use_label_encoder=False, eval_metric='mlogloss'))])
```

```python
# Modelo 3: Regressão Logística
pipeline_lr = Pipeline(steps=[('preprocessor', preprocessor),
                              ('classifier', LogisticRegression(max_iter=1000))])
```

```python
# Dividir em treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Ajustar e avaliar os três modelos
models = [('Random Forest', pipeline_rf), ('XGBoost', pipeline_xgb), ('Logistic Regression', pipeline_lr)]
```

```python
for name, model in models:
  print(f'\nAjustando o modelo: {name}')
  model.fit(X_train, y_train)
  y_pred = model.predict(X_test)
  print(classification_report(y_test, y_pred))
```

- De acordo com os dados apresentados:
 - XGBoost (acurácia de 90%) e Random Forest (acurácia de 91%) aprensetaram os melhores resultados, portanto, os hiperparâmtros destes dois modelos serão otimizados.

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier

# Definir os parâmetros a serem otimizados para Random Forest
param_grid_rf = {
    'classifier__n_estimators': [100, 200, 300],
    'classifier__max_depth': [10, 20, 30],
    'classifier__min_samples_split': [2, 5, 10],
    'classifier__min_samples_leaf': [1, 2, 4],
    'classifier__bootstrap': [True, False]
}

# Usar GridSearchCV para Random Forest
grid_search_rf = GridSearchCV(pipeline_rf, param_grid_rf, cv=5, scoring='accuracy', n_jobs=-1)
grid_search_rf.fit(X_train, y_train)

# Exibir os melhores parâmetros e a acurácia no conjunto de teste
print(f'Melhores parâmetros para Random Forest: {grid_search_rf.best_params_}')
print(f'Acurácia no conjunto de teste (Random Forest): {grid_search_rf.score(X_test, y_test)}')
```

```python
import xgboost as xgb

# Definir os parâmetros a serem otimizados para XGBoost
param_grid_xgb = {
    'classifier__n_estimators': [100, 200, 300],
    'classifier__max_depth': [3, 6, 9],
    'classifier__learning_rate': [0.01, 0.1, 0.3],
    'classifier__subsample': [0.7, 0.8, 1.0],
    'classifier__colsample_bytree': [0.7, 0.8, 1.0]
}

# Usar GridSearchCV para XGBoost
grid_search_xgb = GridSearchCV(pipeline_xgb, param_grid_xgb, cv=5, scoring='accuracy', n_jobs=-1)
grid_search_xgb.fit(X_train, y_train)

# Exibir os melhores parâmetros e a acurácia no conjunto de teste
print(f'Melhores parâmetros para XGBoost: {grid_search_xgb.best_params_}')
print(f'Acurácia no conjunto de teste (XGBoost): {grid_search_xgb.score(X_test, y_test)}')
```

```python
from sklearn.metrics import classification_report, f1_score

# Aplicando o melhor modelo otimizado de Random Forest
best_rf = grid_search_rf.best_estimator_
y_pred_rf = best_rf.predict(X_test)

# Avaliando o modelo Random Forest (com os melhores parâmetros)
print("Random Forest - Avaliação das Métricas:")
print(classification_report(y_test, y_pred_rf))  # Isso inclui o F1-score, precisão, recall e acurácia

# F1-score apenas
f1_rf = f1_score(y_test, y_pred_rf, average='weighted')
print(f'F1-score (Weighted) para Random Forest: {f1_rf}')

# Aplicando o melhor modelo otimizado de XGBoost
best_xgb = grid_search_xgb.best_estimator_
y_pred_xgb = best_xgb.predict(X_test)

# Avaliando o modelo XGBoost (com os melhores parâmetros)
print("\nXGBoost - Avaliação das Métricas:")
print(classification_report(y_test, y_pred_xgb))

# F1-score apenas
f1_xgb = f1_score(y_test, y_pred_xgb, average='weighted')
print(f'F1-score (Weighted) para XGBoost: {f1_xgb}')
```

- Análise:
 - Considerando que o dataset possui variáveis com dados desbalanceados e alguns outliers, decidiu-se levar priorizar o F1-score para escolha do melhor modelo.
 - Ambos, Random Forest e XGBoost, tiverem métricas muito boas e próximas, contudo o primeiro se destacou, apresentando um F1-score de 0,91.
 - Para título de conhecimentos:
  - É uma média ponderada pelo suporte de cada classe. Ele leva em consideração o desbalanceamento de classes, sendo uma métrica muito útil quando há classes minoritárias importantes no conjunto de dados.
